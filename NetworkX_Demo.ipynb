add nodes/edges + visualize them

```python
import networkx as nx
import matplotlib.pyplot as plt
# init network
G = nx.Graph()
# add edges
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(2, 8)
G.add_edge(3, 4)
G.add_edge(2, 5)
G.add_edge(4, 8)
G.add_node("a")        # adds node "spam"
G.add_edge(5, 'a')


# network visualization options
options = {
    "font_size": 36,
    "node_size": 3000,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 6,
    "width": 6,
}
# draw network
nx.draw_networkx(G, **options)

# plot setting and show
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

remove a node

```python
G.remove_node(3)

# draw network
nx.draw_networkx(G, **options)

# plot setting and show
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```
